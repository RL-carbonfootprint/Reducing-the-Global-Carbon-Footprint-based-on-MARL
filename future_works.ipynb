```python
#importing libs
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
#loading dataFrames
LT_df = pd.read_csv('LT_db.csv', index_col=0)
MT_df = pd.read_csv('MT_db.csv', index_col=0)
ST_df = pd.read_csv('ST_db.csv', index_col=0)
LT_df.head()
```

```python
LT_values = LT_df.values[:-1]
Population_LT = LT_df.values[-1]
```

```python
plt.figure(figsize=(16, 8))
plt.scatter(np.arange(0, len(LT_values.ravel())), LT_values.ravel())
plt.legend(['LT group of countries'])
plt.yticks(np.arange(0, 0.1, 0.01))
plt.ylabel('values')
plt.xlabel('CO2 Emission datapoints')
plt.show()
```
